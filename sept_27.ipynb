```python
import numpy as np
from scipy.stats import uniform
```

```python
# random variable support {0,1,2,3}
# p_i = 0.1, 0.3, 0.4, 0.2

def inv(U):
    if U <= 0.1:
        return 0
    elif U <= 0.4:
        return 1
    elif U <= 0.8:
        return 2
    else:
        return 3

def inv_transform(n):
    U = uniform.rvs(size=n)
    X = [inv(u) for u in U]
    return X
```

```python
inv_transform(500)
```

```python
def inverse_transform(n):
    U = uniform.rvs(size=n)
    X = np.ones(n) * 3
    cond0 = U < 0.1
    cond1 = (U <= 0.4) & (U > 0.1)
    cond2 = (U <= 0.8) & (U > 0.4)
    X[cond0] = 0
    X[cond1] = 1
    X[cond2] = 2
    return X
```

```python
inverse_transform(500)
```
